```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# โหลดข้อมูล Features ที่เราสกัดไว้
df = pd.read_csv('../vibration_features.csv') 
# (หมายเหตุ: ถ้าไฟล์ .csv อยู่ข้างนอก notebooks ต้องใช้ ../ เพื่อถอยหลังออกไป 1 step ครับ)

print("ข้อมูลพร้อมสำหรับการวิเคราะห์ครับ!")
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# สร้างกราฟแท่ง
ax = sns.barplot(data=df, x='RMS', y='filename', hue='filename', palette='magma', legend=False)

# ขีดเส้น Threshold อันตราย (ค่า 0.58 ที่เราตั้งไว้ใน Model)
plt.axvline(x=0.58, color='red', linestyle='--', label='Danger Threshold (0.58)')

plt.title('Comparison of Vibration Intensity (RMS)', fontsize=15)
plt.xlabel('RMS Amplitude (G)', fontsize=12)
plt.ylabel('Equipment Data File', fontsize=12)
plt.legend()
plt.show()
```

📝 สรุปผลการพยากรณ์ (Summary of Findings)
จากการวิเคราะห์ข้อมูลความสั่นสะเทือนพบว่า:

Cooling Pump: มีแนวโน้มค่าสั่นสะเทือน (RMS) สูงขึ้นอย่างต่อเนื่องจากเดือน มิ.ย. ถึง ต.ค. และทะลุเกณฑ์ Danger Threshold ในเดือนล่าสุด

Jockey Pump: มีค่าแรงกระแทกสูงสุด (Peak) ที่ค่อนข้างสูง เสี่ยงต่อการเกิดความเสียหายกะทันหัน

ข้อเสนอแนะ: ควรเข้าทำการตรวจสอบ (Maintenance) ของปั๊ม Cooling Pump ทันทีเพื่อป้องกันเครื่องจักรหยุดทำงานนอกแผน (Unplanned Downtime)
